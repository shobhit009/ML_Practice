```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime as dt
from pandas_datareader import data as pdr
yf.pdr_override()
# Code to adjust the date for weekends

startyear = 2013
startmonth = 1
startday = 1

start = dt.datetime(startyear,startmonth,startday)
now = dt.datetime.now()
weekday = now.weekday()
if weekday == 5:
    curr_date = dt.datetime.now() - dt.timedelta(days=1)
elif weekday == 6:
    curr_date = dt.datetime.now() - dt.timedelta(days=2)
else:
    curr_date = dt.datetime.now()

prev_date = dt.datetime.now() - dt.timedelta(days=3)
Master_DF = pd.DataFrame()
Final_DF = pd.DataFrame()
ticker = ["UPL","INFY","TATASTEEL","BAJFINANCE","BAJAJ-AUTO","BPCL","HDFC","IOC",
         "HINDALCO","EICHERMOT","SBILIFE","HINDUNILVR","BAJAJFINSV","ITC","BRITANNIA","RELIANCE",
         "DIVISLAB","TCS","WIPRO","ONGC","TECHM","HDFCBANK","HCLTECH","SUNPHARMA","GRASIM","TITAN",
         "INDUSINDBK","CIPLA","HDFCLIFE","JSWSTEEL","NESTLEIND","KOTAKBANK","NTPC","MARUTI","BHARTIARTL","ADANIPORTS",
         "LT","ULTRACEMCO","ASIANPAINT","TATAMOTORS","M&M","SHREECEM","ICICIBANK","AXISBANK","POWERGRID","COALINDIA",
         "HEROMOTOCO","TATACONSUM","SBIN","DRREDDY",]
```

```python
for i in ticker:
    #Code to get data for a single stock
    i = i + ".NS"
    df = pdr.get_data_yahoo(i,start,curr_date)
    Master_DF = Master_DF.append(df)
    #Calculate RSI for 2 period and 200 ma
    rsi_period = 2

    change = Master_DF['Close'].diff()
    gain = change.mask(change < 0, 0)
    loss = change.mask(change > 0, 0)
    average_gain = gain.ewm(com=rsi_period -1, min_periods=rsi_period).mean()
    average_loss = loss.ewm(com=rsi_period -1, min_periods=rsi_period).mean()
    rs = abs(average_gain/average_loss)
    rsi = 100 - (100/(1+rs))
    Master_DF['rsi'] = rsi
    Master_DF['200MA'] = Master_DF.Close.rolling(200).mean()
    #exclude first 200 days since 200 day moving average will be ineffective
    Master_DF = Master_DF.iloc[200:]
    #process only the latest date of the stock and insert it into the master data
#     print(Master_DF)
    Final_DF = Final_DF.append(Master_DF.iloc[-1])   
    print (Final_DF)
    
#     Final_DF.insert(0, "Test", i)
# print(Final_DF)
# #Check if the stock needs to be shorted
# for i in Final_DF.index:
#     if(((Final_DF['Close'][i] - Final_DF['Open'][i])/ Final_DF['Open'][i])*100 > 3 ) and Final_DF['Close'][i] > Final_DF['200MA'][i] and Final_DF['rsi'][i]>50 :
#         print(i)
```
