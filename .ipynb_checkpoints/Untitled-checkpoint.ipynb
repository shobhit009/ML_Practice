```python
import requests
from datetime import date
import time
import yfinance as yf
from datetime import datetime,timedelta
import os
import json
import csv
import pandas as pd
import numpy as np
# package import statement
import pandas as pd
import arrow
# from market_profile import MarketProfile
signal_data = {}
Master_DF = pd.DataFrame()
prev_date = datetime.now() - timedelta(days=62)
curr_date = datetime.now()
BC = []
TC = []
PP = []

stock = ["UPL","INFY","TATASTEEL","BAJFINANCE","BAJAJ-AUTO","BPCL","HDFC","IOC",
         "HINDALCO","EICHERMOT","SBILIFE","HINDUNILVR","BAJAJFINSV","ITC","BRITANNIA","RELIANCE",
         "DIVISLAB","TCS","WIPRO","ONGC","TECHM","HDFCBANK","HCLTECH","SUNPHARMA","GRASIM","TITAN",
         "INDUSINDBK","CIPLA","HDFCLIFE","JSWSTEEL","NESTLEIND","KOTAKBANK","NTPC","MARUTI","BHARTIARTL","ADANIPORTS",
         "LT","ULTRACEMCO","ASIANPAINT","TATAMOTORS","M&M","SHREECEM","ICICIBANK","AXISBANK","POWERGRID","COALINDIA",
         "HEROMOTOCO","TATACONSUM","SBIN","DRREDDY",]
for ticker in stock:
    signal_data[ticker] = yf.download(ticker+".NS",prev_date,curr_date,interval="1mo")
    my_df_daily = pd.DataFrame(signal_data[ticker])
    my_df_daily.insert(0,'Stock',ticker)
    Master_DF = Master_DF.append(my_df_daily)
    #print(signal_data[ticker])

Master_DF['BC'] = my_df_daily.apply(lambda _: '', axis=1)
Master_DF['PP'] = my_df_daily.apply(lambda _: '', axis=1)
Master_DF['TC'] = my_df_daily.apply(lambda _: '', axis=1)

Master_DF.reset_index(inplace=True)
Master_DF

prev_open = Master_DF['Open']
prev_high = Master_DF['High']
prev_low = Master_DF['Low']
prev_close = Master_DF['Close']

prev_o

Master_DF['PP'] = (prev_high + prev_low + prev_close) / 3
Master_DF['R2'] = Master_DF['PP'] + (prev_high - prev_low)
Master_DF['R1'] = 2*Master_DF['PP'] - prev_low
Master_DF['S1'] = 2*Master_DF['PP'] - prev_high
Master_DF['S2'] = Master_DF['PP'] - (prev_high - prev_low)

Master_DF['BC'] = (prev_high + prev_low)/2

Master_DF['TC'] = (Master_DF['PP'] -Master_DF['BC']) + Master_DF['PP']
Master_DF

for i in Master_DF.index:

    if Master_DF.loc[i,'BC'] > Master_DF.loc[i,'TC']:
        temp = Master_DF.loc[i,'BC']
        Master_DF.loc[i,'BC'] = Master_DF.loc[i,'TC']
        Master_DF.loc[i,'TC'] = temp
#     Master_DF.loc[i,'BC'] = BC_1
#     Master_DF.loc[i,'TC'] = TC_1
#     Master_DF.loc[i,'PP'] = PP
```
