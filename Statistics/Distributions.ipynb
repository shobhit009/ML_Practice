```python
import numpy as np
import random

def roll_dice(n_simulations = 100000):
    count = 0
    
    for i in range(n_simulations):
        die_1 = random.randint(1,6)
        die_2 = random.randint(1,6)
        
        score = die_1 + die_2
        
        if (score % 2 == 0 or score > 7):
            count+=1
            
    return (count/n_simulations)
```

```python
String = 'The probability of rolling an even number or greater than 7 is:'
print(String, np.round(roll_dice()*100,2), '%')
```

```python
import random

trials = 10000
totalA = 12
totalB = 18

success = 0

choiceList = list('A'*totalA + 'B'*totalB)

def transistor():
    return random.choice(choiceList)

for i in range(trials):
    try1 = transistor()
    try2 = transistor()
    
    
    if try1 =='A' and try2 == 'B':
        success += 1
        
print (float(success)/trials)
```

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
plt.ylabel('Probabilty of car passing')
plt.xlabel('Number of cars')
plt.title('Probability Distribution curve')

arr=[]
rv = poisson(25)
for num in range(0,40):
    arr.append(rv.pmf(num))
    
prob = rv.pmf(28)
print(prob)
plt.grid(True)
plt.plot(arr, linewidth=2.0)
plt.plot([28],[prob],marker='o',markersize=6, color='red')
plt.show()
```
